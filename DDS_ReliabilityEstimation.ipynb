```python
from sklearn.preprocessing import StandardScaler

y = df['Reliability']
selected_columns = ['Commits', 'FileChanged', 'Addition', 'Deletion', 'Contributors', 
                    'FileChangedPerCommit', 'Addition/Deletion', 'AdditionPerCommit', 'DeletionPerCommit',
                    'FileChangedPerContributor', 'ElapsedTimeBetweenTwoVersions',
                    'NOM', 'LOC', 'LOC/NOM', 'MVG', 'MVG/NOM', 'COM', 'COM/NOM', 'IF4', 'IF4/NOM', 
                    'WMC', 'DIT', 'NOC', 'CBO']
X = df[selected_columns]
print('Number of features :', len(selected_columns))
print('Features :', selected_columns)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier

min_features_to_select = 1  # Minimum number of features to consider
clf = RandomForestClassifier()
cv = StratifiedKFold(3)

rfecv = RFECV(
    estimator=clf,
    step=1,
    cv=cv,
    scoring="accuracy",
    min_features_to_select=min_features_to_select,
    n_jobs=2,
)
rfecv.fit(X, y)

rfecv_mask = rfecv.get_support()
selected_features = [] 
for bool, feature in zip(rfecv_mask, X.columns):
  if bool:
    selected_features.append(feature)
print('Number of features :', X.shape[1])
print('Optimal number of features :', rfecv.n_features_)
print('Best features :', selected_features)

n_features = X.shape[1]
clf.fit(X, y)
plt.figure(figsize=(8,8))
plt.barh(range(n_features), clf.feature_importances_, align='center') 
plt.yticks(np.arange(n_features), X.columns.values) 
plt.xlabel('Feature importance')
plt.ylabel('Feature')
plt.show()
```

```python
X = df[selected_features]
X = StandardScaler().fit_transform(X)
X.shape
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

# prepare configuration for cross validation test harness
seed = 7

# prepare models
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier(3)))
models.append(('CART', DecisionTreeClassifier(max_depth=5)))
models.append(('RF', RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)))
models.append(('NB', GaussianProcessClassifier(1.0 * RBF(1.0))))
models.append(('SVM', SVC()))
models.append(('NN', MLPClassifier(alpha=1, max_iter=1000)))
models.append(('AdaBoost', AdaBoostClassifier()))

# evaluate each model in turn
results = []
names = []
for name, model in models:
	kfold = model_selection.KFold(n_splits=3)
	accuracy = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
	results.append(accuracy)
	names.append(name)
	msg = "%s => accuracy = %f; std = %f" % (name, accuracy.mean(), accuracy.std())
	print(msg)
```

```python
 # boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
